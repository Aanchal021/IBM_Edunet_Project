```python
# Sales and Profit Analysis

total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
average_sales_per_order = df['Sales'].mean()
average_profit_per_order = df['Profit'].mean()

print("\nSales and Profit Analysis")
print(f"\nTotal Sales: ${total_sales:.2f}")
print(f"Total Profit: ${total_profit:.2f}")
print(f"Average Sales per Order: ${average_sales_per_order:.2f}")
print(f"Average Profit per Order: ${average_profit_per_order:.2f}")
```

```python
# Top Selling Categories and Sub-Categories

top_categories = df.groupby('Category')['Sales'].sum().sort_values(ascending=False).head(5)
top_sub_categories = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(5)

print("\nTop Selling Categories:")
print(top_categories)

print("\nTop Selling Sub-Categories:")
print(top_sub_categories)
```

```python
# Discount and Profit Relationship

discount_profit_relationship = df.groupby('Discount')['Profit'].mean()

print("\nDiscount and Profit Relationship:")
print(discount_profit_relationship)
```

```python
# Region-wise Analysis

region_sales = df.groupby('Region')['Sales'].sum()
region_profit = df.groupby('Region')['Profit'].sum()

print("\nRegion-wise Sales:\n")
print(region_sales)

print("\nRegion-wise Profit:\n")
print(region_profit)
```

```python
# State-wise Analysis

top_states_by_sales = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(5)
bottom_states_by_profit = df.groupby('State')['Profit'].sum().sort_values().head(5)

print("\nTop States by Sales:\n")
print(top_states_by_sales)

print("\nBottom States by Profit:\n")
print(bottom_states_by_profit)
```

```python
# Correlation between Variables

import warnings
warnings.filterwarnings('ignore')

correlation_matrix = df.corr()

print("\nCorrelation Matrix:\n")
print(correlation_matrix)
```

## Data Visualization

#### Bar charts to visualize total sales and profit per product category and total Profit and Quantity sold for each Product Sub-Category.

```python
print(df['Category'].dtype)
```

```python
sns.countplot(x='Category', data=df)
```

Office suplies was the most recurrent among all categories.

```python
# Group the data by the 'Category' column and calculate the total sales from each Category
category_sales = df.groupby('Category')['Sales'].sum().reset_index()

# Creating the bar chart using Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Sales', data=category_sales, palette='viridis')

plt.xticks(rotation=45, ha='right')  # Rotates the x-axis labels for better readability
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.title('Total Sales per Product Category')

plt.tight_layout()
plt.show()
```

```python
# Group the data by the 'Category' column and calculate the total profit for each category
category_profit = df.groupby('Category')['Profit'].sum().reset_index()

# Creating the bar chart using Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Profit', data=category_profit, palette='muted')

plt.xticks(rotation=45, ha='right')  # Rotates the x-axis labels for better readability
plt.xlabel('Product Category')
plt.ylabel('Total Profit')
plt.title('Total Profit per Product Category')

plt.tight_layout()
plt.show()
```

```python
state_sales = df.groupby("State")["Sales"].sum()

# Sort the states based on the sum of sales in descending order and select the top 10
top_10_states = state_sales.sort_values(ascending=False).head(10)
last_10_states = state_sales.sort_values(ascending=False).tail(10)

# Plot the bar chart
fig, axes = plt.subplots(1, 2, figsize=(9, 5))
top_10_states.plot(kind='bar', ax=axes[0])
last_10_states.plot(kind='bar',ax=axes[1])
axes[0].set_title('Top 10 countries sales')
axes[1].set_title('Least 10 countries sales')
plt.xlabel('State')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize= (10,19))
df.groupby('Category')[['Profit', 'Sales']].agg(['sum']).plot.bar()
plt.legend(['Profits','Sales'])

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# First subplot: Discount & Profit relation based on Sub-Category
df.groupby('Sub-Category')[['Discount', 'Profit']].sum().plot(kind='bar', ax=axes[0])
axes[0].set_title('Discount & Profit Relation based on Sub-Category')

# Second subplot: Quantity & Profit relation based on Sub-Category
df.groupby('Sub-Category')[['Quantity', 'Profit']].sum().plot(kind='bar', ax=axes[1])
axes[1].set_title('Quantity & Profit Relation based on Sub-Category')

# plt.xticks(rotation=90)
plt.show()
```

```python
df.groupby('Sub-Category')[['Sales','Quantity']].agg(['sum']).plot(kind='bar')

plt.xticks(rotation=90)
plt.legend(['Sales','Quantity'])
plt.title("Total Sales vs. Quantity ")
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')

# Grouping data by 'Sub-Category' and calculating total profit and total quantity sold
subcategory_profit_quantity = df.groupby('Sub-Category').agg({'Profit': 'sum', 'Quantity': 'sum'}).reset_index()

# Sorting the data based on total profit in descending order to highlight the most profitable sub-categories
subcategory_profit_quantity = subcategory_profit_quantity.sort_values(by='Profit', ascending=False)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Profit', y='Sub-Category', data=subcategory_profit_quantity, palette='coolwarm', ci=None)
plt.xlabel('Total Profit')
plt.ylabel('Product Sub-Category')
plt.title('Total Profit and Quantity Sold for Each Product Sub-Category')
plt.tight_layout()
plt.show()
```

#### Line charts to visualize the trend of total sales over different states and the trend of total profit over different product categories.

```python
# Line chart for total sales trend over different states

state_sales = df.groupby('State')['Sales'].sum().reset_index()

sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.lineplot(x='State', y='Sales', data=state_sales, marker='o', color='b')

plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.title('Total Sales Trend over Different States')
plt.tight_layout()
plt.show()
```

```python
# Line chart for total profit trend over different product categories

category_profit = df.groupby('Category')['Profit'].sum().reset_index()

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

sns.lineplot(x='Category', y='Profit', data=category_profit, marker='o', color='g')

plt.xticks(rotation=45, ha='right')
plt.xlabel('Product Category')
plt.ylabel('Total Profit')
plt.title('Total Profit Trend over Different Product Categories')
plt.tight_layout()
plt.show()
```

#### Pie charts to visualize proportion of total sales contributed by each region and each product category.

```python
# Pie chart for proportion of total sales contributed by each region

region_sales = df.groupby('Region')['Sales'].sum().reset_index()

plt.figure(figsize=(6, 6))
plt.pie(region_sales['Sales'], labels=region_sales['Region'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Proportion of Total Sales by Region')
plt.show()
```

```python
# Pie chart for proportion of total sales contributed by each product category

category_sales = df.groupby('Category')['Sales'].sum().reset_index()

plt.figure(figsize=(6, 6))
plt.pie(category_sales['Sales'], labels=category_sales['Category'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Proportion of Total Sales by Product Category')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
fig.suptitle("Sales with different shipping modes and Segments", fontsize=16)
sns.lineplot(x='Ship Mode', y='Sales', data=df, ax=axes[0])
sns.lineplot(x='Segment', y='Sales', data=df, ax=axes[1])
plt.show()
```

#### Scatter plot charts for: Sales vs. Profit & Quantity vs. Profit

```python
# Scatter plot for Sales vs. Profit

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Sales', palette='cool', edgecolor='black', alpha=0.7, legend=False)

plt.xlabel('Sales')
plt.ylabel('Total Profits')
plt.title('Sales vs. Profit')

plt.grid(True)
plt.show()
```

```python
# Scatter plot for Quantity vs. Profit

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Quantity', y='Profit', hue='Quantity', palette='magma', edgecolor='black', alpha=0.7, legend=False)

plt.xlabel('Quantity')
plt.ylabel('Total Profits')
plt.title('Quantity vs. Profit')

plt.grid(True)
plt.show()
```
